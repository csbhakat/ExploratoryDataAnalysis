## Feature Engineering

In this section we are going to cover following -
1. Handle missing values- Categorical and Numerical
2. Normalize feature
3. Standrise values of the variable

```python
#Here we can get the percentage of missing value in each *categorical* feature.
categorical_features_with_na=[feature for feature in df.columns if df[feature].isnull().values.any()==True and df[feature].dtypes=='O']
for featurename in categorical_features_with_na:
    #print('Percentage of missing value in',featurename,'is',np.round((df[featurename].isnull().sum()/23412)*100,3),'%')
    print('Percentage of missing value in',featurename,'is',np.round((df[featurename].isnull().sum()/23412)*100,3),'%')
```

Magnitude type missing value is being handeled in the below cell.
Since maximum magnitude type is 'MW', therefore we are going to replace the missing value with 'NW'

```python
print('Categories in Magnitude Type \n',df['Magnitude Type'].value_counts())
df['Magnitude Type']=df['Magnitude Type'].fillna('NW')
df['Magnitude Type'].isnull().values.any()
```

```python
#Here we can get the percentage of missing value in each *categorical* feature.
numerical_features_with_na=[feature for feature in df.columns if df[feature].isnull().values.any()==True and df[feature].dtypes!='O']
for featurename in numerical_features_with_na:
    #print('Percentage of missing value in',featurename,'is',np.round((df[featurename].isnull().sum()/23412)*100,3),'%')
    print('Percentage of missing value in',featurename,'is',np.round((df[featurename].isnull().sum()/23412)*100,3),'%')
```

As we can see very large percentage of data is missing in the above mentioned numericals fatures .
